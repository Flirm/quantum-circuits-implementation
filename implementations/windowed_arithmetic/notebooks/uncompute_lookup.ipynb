```python
from qiskit import *
from math import log2, ceil
from qiskit.circuit.library import XGate, UnitaryGate, MCXGate
from qiskit.circuit import Gate
```

```python
def init_unary(in_bits: int) -> QuantumCircuit:
    in_reg = QuantumRegister(in_bits, "input")
    un_reg = QuantumRegister(2**in_bits, "unary")
    quantum_circuit = QuantumCircuit(in_reg, un_reg)

    quantum_circuit.x(un_reg[0])
    for i in range(in_bits):
        n = 2**i
        for j in range(n):
            quantum_circuit.cswap(in_reg[i], un_reg[j], un_reg[j+(2**i)])
    return quantum_circuit

init_unary(3).draw("mpl")
```
